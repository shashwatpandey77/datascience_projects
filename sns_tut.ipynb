```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
data['time'].value_counts().plot(kind='pie')
```

**Relational Plot (Relplot)**

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='sex')
```

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex',size='size',sizes=(15,200))
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='sex')
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,hue='region',style='event',linewidth=2,ci=False)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri.query("event=='stim'"),linewidth=2,ci=False)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,row='region',col='event')
```

**Categorical Plot (Catplot)**

```python
sns.catplot(x='day',y='tip',data=data,kind='strip')
```

```python
sns.catplot(x='day',y='tip',data=data,kind='swarm')
```

```python
sns.swarmplot(x='day',y='tip',data=data)
```

**Box Plot (Distribution insights)**

```python
sns.boxplot(data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='box')
```

**Barplot and Countplot**

```python
sns.catplot(x='sex',y='tip',data=data,hue='smoker',kind='bar')
```

```python
sns.catplot(x='sex',data=data,hue='smoker',kind='count')
```

```python
p_data = data.pivot_table(index='size',columns='day',values='total_bill',aggfunc='sum')
```

```python
p_data
```

```python
sns.heatmap(p_data,linewidth=0.5,cmap='flare')
```

```python
sns.histplot(x='total_bill',data=data,element='poly') #bars and step
```

**Distplot**

```python
titan = sns.load_dataset('titanic')
```

```python
titan.head()
```

```python
titan['age'].fillna(titan['age'].mean(),inplace=True)
```

```python
sns.displot(titan['age'],kind='kde')
```

```python
sns.kdeplot(titan['age'])
```

```python
sns.kdeplot(titan[titan['survived']==1]['age'])
sns.kdeplot(titan[titan['survived']==0]['age'])
```
