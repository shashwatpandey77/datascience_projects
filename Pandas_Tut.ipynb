**Most Destructive Batsmen in Death Overs (15-20)**
Should have played 200 balls


```python
msk1 = dl['over']>=15
msk2 = dl['over']<=20
dthovers = dl[msk1 & msk2]
```

```python
allstrikers = dthovers.groupby('batsman')['ball'].count()
x = allstrikers>200
strikerslist = allstrikers[x].index.tolist()
```

```python
strikers = dthovers[dthovers['batsman'].isin(strikerslist)]
```

```python
runs = strikers.groupby('batsman')['batsman_runs'].sum()
balls = strikers.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100 
```

```python
dest = sr.sort_values(ascending=False).reset_index()
```

```python
dest.rename(columns={'batsman_runs':'strike_rate'})
```

**Merge Practice**

```python
temp = pd.DataFrame({'city':['NY','LA','SF'],'temp':[21,14,35]})
humid = pd.DataFrame({'city':['NY','DC','LV','SF'],'humid':[30,45,50,60]})
```

```python
temp
```

```python
temp['diff'] = temp['temp'].map(lambda x : temp['temp'].mean()-x)
```

```python
temp
```

```python
humid
```

```python
temp.merge(humid, on='city') #by default inner (intersection)
```

```python
temp.merge(humid, on='city',how='outer', indicator=True) #outer means union
```

```python
temp.merge(humid, on='city',how='right') #values from right df only
```

```python
temp.rename(columns={'temp':'temprature','diff':'difference'},inplace=True)
```

```python
th = temp.merge(humid,on='city')
```

```python
th['sum']=th['temprature'].map(lambda x : x*2 -5) + th['humid']
```

```python
th
```

```python
th['div']=th['temprature']/th['humid']
th
```
